```python
import pygame
import random

pygame.init()
screen = pygame.display.set_mode((400, 400))
pygame.display.set_caption("Падающие яйца")
font = pygame.font.SysFont(None, 30)

background = pygame.image.load("pnr.jpg")
background = pygame.transform.scale(background, (400, 400))
key_sound = pygame.mixer.Sound("egg.wav")
korzina = pygame.image.load("korzina.png").convert()
korzina.set_colorkey((0, 0, 0))
korzina = pygame.transform.scale(korzina, (80, 80))
picofegg = pygame.image.load("picofegg.jpg").convert()
picofegg.set_colorkey((255,255,255))
picofegg = pygame.transform.scale(picofegg, (40, 40))

spawn_points = [(50, 50), (150, 50), (250, 50), (350, 50)]

running = True
clock = pygame.time.Clock()

balls = []  # список активных шаров (координаты)
spawn_interval = 2000  
last_spawn_time = 0
#speed of korzinka
xk = 200
yk = 400
xk_speed = 0
#kasaniye
player_radius = 35
target_radius = 20
score = 0
death = 0

while running:
    current_time = pygame.time.get_ticks()
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_a:
                    xk_speed = -0.05
                elif event.key == pygame.K_d:
                    xk_speed = +0.05
    
    xk+=xk_speed
    if xk + korzina.get_width() // 2 >= 400:
        xk = 400 - korzina.get_width() // 2
        xk_speed = 0
    if xk - korzina.get_width() // 2 <= 0:
        xk = korzina.get_width() // 2
        xk_speed = 0





    screen.blit(background, (0, 0))
    screen.blit(korzina, (int(xk) - korzina.get_width() // 2, yk - korzina.get_height() // 2 - 20))


    if current_time - last_spawn_time > spawn_interval:
        x, y = random.choice(spawn_points)
        balls.append([x, y])
        last_spawn_time = current_time

    adda = ['0', '1']
    ad = random.choice(adda)
    for ball in balls:
        
        if ad == '0':
            ball[0] -= 0.1
        else:
            ball[0] += 0.1    
        ball[1] += 0.01  # пикселей в миллисекунду

    for ball in balls:
        target_x,target_y = ball[0], ball[1]
        player_x,player_y = int(xk) - korzina.get_width() // 2, yk - korzina.get_height() // 2 - 20
        dx = player_x - target_x
        dy = player_y - target_y
        distance = (dx**2 + dy**2)**0.5
        if distance < player_radius + target_radius:
            score += 1
            ball[1] += 1000
        if ball[1]>=400 and ball[1]<1000:
            death += 1
            key_sound.play()

    # Убираем шары, которые упали ниже 
    balls = [ball for ball in balls if ball[1] < 400]
    
    
    if death>=3:
        pygame.quit()

    
    for x, y in balls:
        screen.blit(picofegg, (x, y))

    
    text = font.render(f"Score: {score}", True, (255, 0, 0))
    screen.blit(text, (10, 10))
    
    pygame.display.flip()
    

pygame.quit()
```
