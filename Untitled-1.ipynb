```python
import turtle

screen = turtle.Screen()
screen.bgcolor("#87CEEB")

t = turtle.Turtle()
t.speed(3)
t.pensize(3)

def draw_rectangle(t, width, height, color_fill, color_pen="black"):
    t.color(color_pen, color_fill)
    t.begin_fill()
    for _ in range(2):
        t.forward(width)
        t.left(90)
        t.forward(height)
        t.left(90)
    t.end_fill()

def draw_triangle(t, side, color_fill, color_pen="black"):
    t.color(color_pen, color_fill)
    t.begin_fill()
    for _ in range(3):
        t.forward(side)
        t.left(120)
    t.end_fill()

def draw_smoke(t, x, y, radius, count):
    t.color("#B0B0B0", "#D3D3D3")
    for i in range(count):
        t.penup()
        t.goto(x + i*15, y + i*20)
        t.pendown()
        t.begin_fill()
        t.circle(radius)
        t.end_fill()

t.penup()
t.goto(-150, -100)
t.pendown()
draw_rectangle(t, 300, 200, "#8B4513")

t.penup()
t.goto(-170, 100)
t.pendown()
draw_triangle(t, 340, "#A52A2A")

t.penup()
t.goto(-40, -100)
t.pendown()
draw_rectangle(t, 80, 120, "#654321")

t.penup()
t.goto(30, -60)
t.pendown()
t.color("gold")
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(-130, 0)
t.pendown()
draw_rectangle(t, 60, 60, "#ADD8E6")
t.color("black")
t.penup()
t.goto(-130, 30)
t.pendown()
t.forward(60)
t.penup()
t.goto(-100, 0)
t.pendown()
t.left(90)
t.forward(60)
t.right(90)

t.penup()
t.goto(70, 0)
t.pendown()
draw_rectangle(t, 60, 60, "#ADD8E6")
t.color("black")
t.penup()
t.goto(70, 30)
t.pendown()
t.forward(60)
t.penup()
t.goto(100, 0)
t.pendown()
t.left(90)
t.forward(60)
t.right(90)

t.penup()
t.goto(80, 140)
t.pendown()
draw_rectangle(t, 40, 80, "#555555")

draw_smoke(t, 110, 220, 15, 5)

t.hideturtle()
turtle.done()
```
